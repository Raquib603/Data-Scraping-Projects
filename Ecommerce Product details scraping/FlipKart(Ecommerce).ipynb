```python
import requests

url = "https://www.flipkart.com/audio-video/headset/headphones/wireless-headphones/pr?sid=0pm,fcn,gc3,ka8"
headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/124.0 Safari/537.36",
}

response = requests.get(url, headers=headers)

if response.status_code == 200:
    with open("flipkart_page.html", "w", encoding="utf-8") as f:
        f.write(response.text)
    print("Saved as flipkart_page.html")
else:
    print("Failed:", response.status_code)
```

```python
from bs4 import BeautifulSoup
import pandas as pd

# Read the saved HTML file
with open("flipkart_page.html", "r", encoding="utf-8") as f:
    html = f.read()

soup = BeautifulSoup(html, "lxml")

# Lists to store data
product_names = []
product_prices = []
product_ratings = []

# Update product selector: Flipkart products are often in divs with a 'data-id' attribute.
# This is a more robust way to identify individual product cards.
products = soup.find_all("div", attrs={"data-id": True})

for product in products:
    # Product name
    name_tag = product.find("a", class_="wjcEIp")
    if name_tag and "title" in name_tag.attrs:
        product_names.append(name_tag["title"])
    else:
        product_names.append("N/A") # Ensure lists stay aligned even if a product lacks a name

    # Product price
    price_tag = product.find("div", class_="Nx9bqj")
    if price_tag:
        product_prices.append(price_tag.text.replace("â‚¹", "").replace(",", "").strip())
    else:
        product_prices.append("N/A")

    # Product rating
    rating_tag = product.find("div", class_="XQDdHH")
    if rating_tag:
        product_ratings.append(rating_tag.text.strip())
    else:
        product_ratings.append("N/A")

# Create DataFrame
df = pd.DataFrame({
    "Name": product_names,
    "Price": product_prices,
    "Rating": product_ratings
})

# Save to CSV
df.to_csv("flipkart_headphones.csv", index=False)
print("Saved as flipkart_headphones.csv")
```
